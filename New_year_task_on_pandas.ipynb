```python
class Car261_264:
    def __init__(self, path):
        self.df = pd.read_csv(path)
    
    
    def assert_columns(self):
        class_columns=['id_1', 'id_2', 'route', 'moto', 'car','rv', 'bus', 'truck']
        data_columns=list(self.df.columns)
        assert data_columns==class_columns, "Columns do not match the expected formart" 

    
    def Car_matrix_generation(self):
        df=pd.read_csv('C:\\Users\\pc\\Batch 261\\Pandas\\dataset-1.csv')
        matrix=df.pivot_table(index='id_1', columns='id_2', values='car')
        matrix=matrix.fillna(0)
        return matrix
    
    
    def Car_type_count(self):
        df=pd.read_csv('C:\\Users\\pc\\Batch 261\\Pandas\\dataset-1.csv')
        def foo(val):
            types=[]
            if val<=15:
                return 'low'
            elif 16<=val<= 25:
                return 'medium' 
            else: 
                return 'high'
        df['car_type'] = df['car'].apply(foo)
        type_counts = df['car_type'].value_counts().to_dict()
        type_counts = dict(sorted(type_counts.items()))
        return type_counts
    
    
    def Bus_count_indexes(self):
        df = pd.read_csv('C:\\Users\\pc\\Batch 261\\Pandas\\dataset-1.csv')
        bus_mean = df['bus'].mean()
        condition = df['bus'] > 2*bus_mean
        indices = df[condition].index.tolist()
        indices.sort()
        return indices
    
    
    def Route_filtering(self):
        df = pd.read_csv('C:\\Users\\pc\\Batch 261\\Pandas\\dataset-1.csv')
        route_group = df.groupby('route')['truck'].mean()
        filtered = route_group[route_group > 7].index.tolist()
        filtered.sort()
        return filtered
    
    
    def Matrix_multiply(self,df1):
        df = pd.read_csv('C:\\Users\\pc\\Batch 261\\Pandas\\dataset-1.csv') 
        df[df > 20] = df[df > 20] * 0.75
        df[df <= 20] = df[df <= 20] * 1.25
        df = df.round(1)
        return df
  
```
