## Give a detailed data about the Tamilnadu, and then tell me how many items were sold in total from this warehouse over the 10 year period?

```python
tamilnadu_data = df.loc['Tamilnadu']
tamilnadu_data
```

```python
total_items_sold = tamilnadu_data.sum().sum()
print(f"\nTotal items sold from Tamilnadu warehouse over the 10-year period: {total_items_sold}")
```

## How many Smart TV were sold in Telangana & Karnataka during the year 2017?

```python
smart_tv_sales_2017 = df.loc[(slice(None), slice(None), 'Smart TV'), ('2017', slice(None))]

smart_tv_sales_telangana = smart_tv_sales_2017.loc['Telangana'].sum().sum()
smart_tv_sales_karnataka = smart_tv_sales_2017.loc['Karnataka'].sum().sum()

print(f"Number of Smart TVs sold in Telangana during 2017: {smart_tv_sales_telangana}")
print(f"Number of Smart TVs sold in Karnataka during 2017: {smart_tv_sales_karnataka}")
```

## Get me list of all items that were sold in the Jul-Dec semester of the year 2020 for all States

```python
items_sold_2020_jul_dec = df.loc[:, ('2020', 'Jul-Dec')]
items_list_2020_jul_dec = items_sold_2020_jul_dec.index.get_level_values('Product').unique()

print("List of all items sold in the Jul-Dec semester of the year 2020 for all states:")
print(items_list_2020_jul_dec)
```
