```python
def Neuron(a=None, b=""):
    if a is None:
        a = []
    
    if not all(x in {0, 1} for x in a):
        return "enter correct values"
    
    operations = {
        "and": all,
        "or": any,
        "nor": lambda x: not any(x)
    }
    
    if b not in operations:
        return "invalid operation"
    
    return int(operations[b](a))
```

```python
z = (Neuron([1, 1, 1], "and"))  
x = (Neuron([1, 1, 0], "or"))  
print(z)
print(x)
```

```python
g = (Neuron([0,1,1,1], "is"))
g
```
